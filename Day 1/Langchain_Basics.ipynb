## What is LangChain?

LangChain is an open-source framework for developing applications powered by language models. It provides a standard interface for chains, which are sequences of calls to language models or other utilities. LangChain makes it easy to:

*   **Connect LLMs to external data:** Bring your own data to the LLM.
*   **Combine LLMs with other components:** Build complex applications by chaining together different components.
*   **Develop agentic applications:** Create applications where the LLM can interact with its environment.

In essence, LangChain helps you move beyond simple prompt-response interactions and build more sophisticated LLM-powered applications.

LangChain simplifies and standardizes tasks like:

>Prompt management and optimization

>Chaining multiple LLM calls together

>Integrating external data (RAG: Retrieval-Augmented Generation)

>Building chatbots, agents, and workflows

LangChain supports various LLM providers like OpenAI, Anthropic, Cohere, and now also Google Gemini (via Google Generative AI SDK).

ðŸ§± Step 2: Initialize Gemini Model with LangChain

```python
from langchain_google_genai import ChatGoogleGenerativeAI

# Load Gemini 1.5 Flash model via LangChain wrapper
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)
```

ðŸ“¦ Step 3: Use Basic Prompt to Summarize Text

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Define a prompt template
prompt = ChatPromptTemplate.from_template(
    "Summarize the following text in simple language:\n\n{text}"
)

# Connect prompt -> model -> output parser
chain = prompt | llm | StrOutputParser()

# Example input
input_text = """
LangChain is an open-source framework that simplifies the development of applications using large language models.
It provides tools for chaining LLM calls, integrating with external data sources, and creating intelligent agents.
"""

# Run the chain
summary = chain.invoke({"text": input_text})
print(summary)
```
